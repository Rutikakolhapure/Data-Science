# Bar plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var = sns.load_dataset("penguins")
var
```

```python
order_1=["Adelie", "Chinstrap", "Gentoo"]
sns.barplot(x="species", y="bill_length_mm", data=var,hue="sex",order=order_1,hue_order=["Female","Male"],errorbar=('ci', 100),n_boot=2,orient="v")
plt.show()
```

```python
order_1=["Adelie", "Chinstrap", "Gentoo"]
sns.barplot(x="bill_depth_mm", y="bill_length_mm",data=var,orient="h",palette="icefire",saturation=0.5,ci="sd",hue="species",hue_order=order_1,errorbar=('ci', 100),n_boot=2)
plt.show()
```

```python
order_1=["Adelie", "Chinstrap", "Gentoo"]
sns.barplot(x="species", y="bill_length_mm", data=var,hue="sex",order=order_1,hue_order=["Female","Male"],errorbar=('ci', 100),n_boot=2,orient="v"
            ,errwidth=10,errcolor="r")
plt.show()
```
