# Line Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

var=[1,2,3,4,5,6,7]
var_1=[2,3,4,1,5,6,7]

x=pd.DataFrame({"var":var,"var_1":var_1})

sns.lineplot(x="var",y="var_1",data=x)
plt.show()
```
