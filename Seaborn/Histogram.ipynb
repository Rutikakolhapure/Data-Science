# Histogram Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var = sns.load_dataset("penguins")
var
```

```python
sns.displot(var["flipper_length_mm"],bins=[170,180,190,200,210,220,230,240],kde=True,rug = True,color="red",height=5,aspect=1.5)
plt.show()
```

```python
sns.kdeplot(data=var['bill_length_mm'],shade=True)
```
