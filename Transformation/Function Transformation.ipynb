```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv(r'C:\Users\ruchi\Videos\dataset\DataSets_For_Learning\loan.csv')
dataset.head(4) 
```

```python
sns.distplot(dataset["CoapplicantIncome"])
```

```python
q1 = dataset["CoapplicantIncome"].quantile(0.25)
q3 = dataset["CoapplicantIncome"].quantile(0.75)
iqr=q3-q1
```

```python
min_range = q1 - 1.5 * iqr
max_range = q3 + 1.5 * iqr
min_range, max_range
```

```python
dataset[dataset["CoapplicantIncome"] <= max_range] 
```

```python
dataset = dataset[dataset["CoapplicantIncome"] <= max_range] 
```

```python
from sklearn.preprocessing import FunctionTransformer
```

```python
ft = FunctionTransformer(func = np.log1p)
```

```python
ft.fit_transform(dataset[["CoapplicantIncome"]])
```

```python
dataset["CoapplicantIncome_ft"]=ft.fit_transform(dataset[["CoapplicantIncome"]])
```

```python
sns.distplot(dataset["CoapplicantIncome_ft"])
```
