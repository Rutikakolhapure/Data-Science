### Normalization Min-Max Method

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv(r'C:\Users\ruchi\Videos\dataset\DataSets_For_Learning\loan.csv')
dataset.head(4)  # Display the first 4 rows of the dataset
```

```python
dataset.describe()
```

```python
dataset.isnull().sum()
```

```python
sns.distplot(dataset["CoapplicantIncome"])
```

```python
from sklearn.preprocessing import MinMaxScaler 
```

```python
ms = MinMaxScaler()
ms.fit(dataset[["CoapplicantIncome"]])
```

```python
ms.transform(dataset[["CoapplicantIncome"]])
```

```python
dataset["CoapplicantIncome_min"] = ms.transform(dataset[["CoapplicantIncome"]])
```

```python
dataset.head(3)
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title("Before")
sns.distplot(dataset["CoapplicantIncome"])

plt.subplot(1,2,2)
plt.title("After")
sns.distplot(dataset["CoapplicantIncome_min"])
```
