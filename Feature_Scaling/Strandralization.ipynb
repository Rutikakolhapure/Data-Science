#### Standardization

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv(r'C:\Users\ruchi\Videos\dataset\DataSets_For_Learning\loan.csv')
dataset.head(4)  # Display the first 4 rows of the dataset
```

```python
dataset.isnull().sum()
```

```python
dataset.fillna({"ApplicantIncome": dataset["ApplicantIncome"].mean()}, inplace=True)
```

```python
sns.distplot(dataset["ApplicantIncome"])
```

```python
dataset.describe() 
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
ss.fit(dataset[["ApplicantIncome"]])
```

```python
ss.transform(dataset[["ApplicantIncome"]])
```

```python
dataset["ApplicantIncome_ss"] = pd.DataFrame(ss.transform(dataset[["ApplicantIncome"]]),columns=['x'])
```

```python
dataset
```

```python
dataset.describe()
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1,2,1)
plt.title("Before")
sns.distplot(dataset["ApplicantIncome_ss"])

plt.subplot(1,2,2)
plt.title("After")
sns.distplot(dataset["ApplicantIncome"])

plt.show()
```
