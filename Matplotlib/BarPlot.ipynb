# Bar Plot

```python
import matplotlib.pyplot as plt

x=["python","C","C++","Java"]
y=[85,70,60,82]
plt.xlabel("languages",fontsize=15)
plt.ylabel("Numbers",fontsize=15)
plt.title("Basic Example of Bar plot:",fontsize=20)
plt.bar(x,y,width=0.4,color="Pink")
plt.show()
```

```python
import matplotlib.pyplot as plt

x=["python","C","C++","Java"]
y=[85,70,60,82]
plt.xlabel("languages",fontsize=15)
plt.ylabel("Numbers",fontsize=15)
plt.title("Basic Example of Bar plot:",fontsize=20)
c=["y","b","r","g"]
plt.bar(x,y,width=0.4,color=c,align="edge",edgecolor="b",linewidth=5,linestyle=":",alpha=0.5) #align="center"
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=["python","C","C++","Java"]
y=[85,70,60,82]
z=[20,30,40,50]

width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]


plt.xlabel("languages",fontsize=15)
plt.ylabel("Numbers",fontsize=15)
plt.title("Basic Example of Bar plot:",fontsize=20)

plt.bar(p,y,width,color="Pink",label="Population")
plt.bar(p1,z,width,color="Black",label="Women")

plt.xticks(p+width/2,x,rotation=20)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

x=["python","C","C++","Java"]
y=[85,70,60,82]
plt.xlabel("languages",fontsize=15)
plt.ylabel("Numbers",fontsize=15)
plt.title("Basic Example of Bar plot:",fontsize=20)
c=["y","b","r","g"]
width=0.4
plt.barh(x,y,width,color=c,align="edge",edgecolor="b",linewidth=5,linestyle=":",alpha=0.5) #align="center"
plt.show()
```
