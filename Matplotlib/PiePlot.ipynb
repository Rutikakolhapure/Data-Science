# Pie Plot

```python
import matplotlib.pyplot as plt

x=[10,20,30,40]
y=["c","c++","Java","Python"]
ex=[0.3,0.0,0.0,0.0]
c=["r","b","g","y"]

plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.1f%%",shadow=True,radius=1,startangle=180,
        textprops={"fontsize":15},counterclock=False,wedgeprops={"linewidth":1,"edgecolor":"Black"},rotatelabels=False
        ,center=(3,3))

plt.title("WScubetecgh")
plt.legend(loc=2)
plt.show()
```

```python
plt.pie([1])
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[10,20,30,40]
x1=[40,30,20,10]
y=["c","c++","Java","Python"]

c=["b","m","Black","y"]
plt.pie(x,labels=y,radius=1.5)
plt.pie(x1,radius=1,colors=c)
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[10,20,30,40]
x1=[40,30,20,10]
y=["c","c++","Java","Python"]

c=["b","r","m","y"]
plt.pie(x,labels=y,radius=1.5)
plt.pie([1],colors="w")
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[10,20,30,40]
x1=[40,30,20,10]
y=["c","c++","Java","Python"]

c=["b","r","m","y"]
plt.pie(x,labels=y,radius=1.5)
cr=plt.Circle(xy=(0,0),radius=1,facecolor="w")
plt.gca().add_artist(cr)

plt.show()
```
