# Scatter Plot

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6,7]
y=[2,3,4,5,6,7,8]
plt.scatter(x,y,color="g")
plt.title("Scatter Plot Example",fontsize=15)
plt.xlabel("Day")
plt.ylabel("Number")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6,7]
y=[2,3,4,5,6,7,8]
c=["r","b","r","b","r","b","r"]
size=[50,40,300,50,60,40,100]
plt.scatter(x,y,color=c,s=size,alpha=0.8,marker="*",edgecolor="g",linewidth=2)
plt.title("Scatter Plot Example",fontsize=15)
plt.xlabel("Day")
plt.ylabel("Number")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6,7]
y=[2,3,4,5,6,7,8]
y1=[5,6,8,6,3,1,9]
colors=[10,40,30,29,56,20,30]
size=[50,40,300,50,60,40,100]
plt.scatter(x,y,c=colors,s=size,cmap="Purples")
plt.scatter(x,y1,c=colors,s=size,cmap="RdBu")
t=plt.colorbar()
t.set_label("ColorBar")
plt.title("Scatter Plot Example",fontsize=15)
plt.xlabel("Day")
plt.ylabel("Number")
plt.show()
```
