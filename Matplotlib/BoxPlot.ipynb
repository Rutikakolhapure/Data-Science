# Box Plot

```python
import matplotlib.pyplot as plt

x=[10,20,30,40,50,60,70,120,200]

plt.boxplot(x,notch=True,vert=True,widths=0.2,labels=["python"],patch_artist=True,
            showmeans=True,sym="g+",boxprops=dict(color="r"),capprops=dict(color="r"),whiskerprops=dict(color="r"),flierprops=dict(markeredgecolor="y")) #whis=4 for joining with outler

plt.show()
```

```python
import matplotlib.pyplot as plt

x=[10,20,30,40,50,60,70]
x1=[10,20,30,40,50,60,70]

y=[x,x1]
plt.boxplot(y,notch=False,labels=["python","c++"],vert=True,widths=0.2,patch_artist=False,
            showmeans=True,sym="g+",boxprops=dict(color="r"),capprops=dict(color="r"),whiskerprops=dict(color="r"),flierprops=dict(markeredgecolor="y")) #whis=4 for joining with outler


plt.show()
```
