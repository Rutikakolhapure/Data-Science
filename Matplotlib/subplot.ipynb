# SubPlot

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[2,3,4,5,6]
plt.subplot(2,2,1)  # 2 rows, 2 columns, first subplot
plt.plot(x,y)

plt.subplot(2,2,2)  # 2 rows, 2 columns, second subplot
plt.pie([1],colors="g")

x1=[10,20,30,40]
plt.subplot(2,2,3)  # 2 rows, 2 columns, third subplot
plt.pie(x1)

x2=["a","s","d","f"]
y2=[10,20,30,40]
plt.subplot(2,2,4)  # 2 rows, 2 columns, fourth subplot
plt.bar(x2,y2)

plt.savefig("line6",facecolor="Pink",transparent=False,bbox_inches="tight")
plt.show()
```
