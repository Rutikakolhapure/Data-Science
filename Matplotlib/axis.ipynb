# Text

```python
import matplotlib.pyplot as plt

x= [1, 2, 3, 4, 5]
y= [2, 3, 5, 7, 11]

plt.plot(x,y)

plt.text(2.5,4,"Data Science", fontsize=15, color='red',style='italic',bbox=dict(facecolor='yellow', alpha=0.5))

plt.annotate('This is a point', xy=(3, 5), xytext=(4, 6),
             arrowprops=dict(facecolor='black', shrink=0.0005))
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('My First Plot')
plt.legend("up",loc=9,facecolor="pink",edgecolor="brown",framealpha=0.2)
plt.show()
```
