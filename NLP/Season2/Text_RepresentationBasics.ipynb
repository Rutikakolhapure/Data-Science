Representing text as a vector is also known as <b>Vector Space Model</b>

Approches of converting text into vector:
1. One Hot Encodding
2. Bag of Words
3. TF-IDF
4. Word Embeddings

Often in NLP, feeding a good text representation to an ordinary algorithm will get you much farther compared to applying top-notch algorithm to an oridinary text representation.
