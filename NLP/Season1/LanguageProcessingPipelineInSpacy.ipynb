tagger Components give PRON: Proper nouns. and lemmitizer gives lemma

<li>token.pos_:Parts of speech </li>
<li>token.lemma_: actual word of the perticular 
word like ate->eat</li>

```python
doc = nlp("Captain America ate 100$ of samosa. Then he said I can do this all day.")

for token in doc:
    print(token, " |", token.pos_, " | ", token.lemma_ )  
```

<li>ent: Entity</li>
<li>spacy.explain : explains exactly what it is.</li>

```python
doc = nlp("Tesla Inc is going to acquire twitter for $45 billion")

for ent in doc.ents:
    print(ent, " | ", ent.label_, " | ", spacy.explain(ent.label_))
```

```python
from spacy import displacy

displacy.render(doc, style="ent", jupyter=True)
```

```python
nlp = spacy.blank("en")

doc = nlp("Tesla Inc is going to acquire twitter for $45 billion")

for ent in doc.ents:
    print(ent, " | ", ent.label_, " | ", spacy.explain(ent.label_))
```

As you can see no output because the piplien is empty

Hence customizing the blank pipeline

```python
source_nlp = spacy.load("en_core_web_sm")

nlp = spacy.blank("en")

nlp.add_pipe("ner",source = source_nlp)
nlp.pipe_names
```

```python
doc = nlp("Tesla Inc is going to acquire twitter for $45 billion")

for ent in doc.ents:
    print(ent, " | ", ent.label_, " | ", spacy.explain(ent.label_))
```
