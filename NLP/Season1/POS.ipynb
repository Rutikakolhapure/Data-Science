Parts Of Speech

```python
import spacy
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
doc = nlp("Elon flew to mars yesterday. He carried biryani masala with him")
for token in doc:
    print(token," | ", token.pos_," | ", spacy.explain(token.pos_))
```

```python
doc = nlp("Wow! Dr. Strange made 265 million $ on the very first day")

for token in doc:
    print(token," | ", token.pos_," | ", spacy.explain(token.pos_)," | ", token.tag_, " | ", spacy.explain(token.tag_))
```

```python
doc = nlp("He quits the job")

doc[1]
print(doc[1].text, " | ", doc[1].tag_, " | ", spacy.explain(doc[1].tag_))
```

```python
earning_text = """ Revenue increased $36.6 billion or 15% with growth across each of our segments etc. Intelligent Cloud revenue increased driven by Azure. Productivity and Business Processes revenue increased driven by Microsoft 365 Commercial cloud. More Personal Computing revenue increased driven by Gaming and Search and news advertising.Cost of revenue increased $13.7 billion or 19% driven by growth in Microsoft Cloud.Gross margin increased $22.9 billion or 13% with growth across each of our segments.•Gross margin percentage decreased slightly driven by Intelligent Cloud, offset in part by More Personal Computing.•Microsoft Cloud gross margin percentage decreased to 69% driven by the impact of scaling our AI infrastructure, offset in part by efficiency gains in Azure.Operating expenses increased $3.8 billion or 6% driven by investments in cloud and AI engineering and Gaming, including the impact of the Activision Blizzard acquisition.Operating income increased $19.1 billion or 17% with growth across each of our segments.
"""
```

```python
doc = nlp(earning_text)

for token in doc:
    print(token," | ", token.pos_," | ", spacy.explain(token.pos_)," | ", token.tag_, " | ", spacy.explain(token.tag_))
```

```python
doc = nlp(earning_text)

for token in doc:
    if token.pos_ in ["SPACE","X","PUNCT"]:
        print(token, " | ", token.pos_, " | ", spacy.explain(token.pos_))
```

```python
doc = nlp(earning_text)

filtered_tokens = []

for token in doc:
    if token.pos_ in ["SPACE","X","PUNCT"]:
        filtered_tokens.append(token)
```

```python
filtered_tokens[:20]
```

```python
count = doc.count_by(spacy.attrs.POS)
count
```

```python
for k,v in count.items():
    print(doc.vocab[k].text," | ",v)
```

```python

import spacy
nlp = spacy.load("en_core_web_sm")
```

```python

news_text = """Inflation rose again in April, continuing a climb that has pushed consumers to the brink and is threatening the economic expansion, the Bureau of Labor Statistics reported Wednesday.\n\nThe consumer price index, a broad-based measure of prices for goods and services, increased 8.3% from a year ago, higher than the Dow Jones estimate for an 8.1% gain. That represented a slight ease from Marchâ€™s peak but was still close to the highest level since the summer of 1982.\n\nRemoving volatile food and ene"""
```

```python
doc = nlp(news_text)

numeral_tokens = []
noun_tokens = []

for token in doc:
    if token.pos_ == "NOUN":
        noun_tokens.append(token)
    elif token.pos_ == 'NUM':
        numeral_tokens.append(token)
```

```python

numeral_tokens[:10]
```

```python
noun_tokens[:10]
```

```python

count = doc.count_by(spacy.attrs.POS)
count
```

```python

for k,v in count.items():
    print(doc.vocab[k].text, "|",v)
```
