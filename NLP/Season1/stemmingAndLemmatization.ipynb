<li>Stemming:
Use fixed rules such as remove able, ing etc. to derive a base word.
ex: ability-> abil which is not correct.
but still it is used in NLTK not in spacy</li>

<li>Lemmatization:
Use knowledge of a language(a.k.a linguistic knowledge) to derive a base word</li>

```python
import nltk
import spacy
```

Stemming in nltk

```python
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
```

```python
words = ["eating","eats","eat","ate","adjustable","rafting","ability","meeting"]

for word in words:print(word, "|", stemmer.stem(word))
```

lemmtization in spacy

```python
nlp = spacy.load("en_core_web_sm")
doc = nlp("eating eats eat ate adjustable rafting ability meeting better")

for token in doc:
    print(token, "|", token.lemma_, "|", token.lemma)
```

```python
doc = nlp("Mando talked for 3 hours although talking isn't his thing he became talkative")
for token in doc:
    print(token, "|", token.lemma_)
```

```python
nlp.pipe_names
```

```python
ar = nlp.get_pipe("attribute_ruler")

ar.add([[{"TEXT":"Bro"}],[{"TEXT":"Brah"}]],{"LEMMA":"Brother"})

doc = nlp("Bro, you wanna go? Brah, don't say no! I am exhausted")
for token in doc:
    print(token.text,"|",token.lemma_)
```
