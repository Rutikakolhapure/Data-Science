1. NLTK is String processing and Spacy is object oriented

2. Spacy Provides efficient NLP algo for the given task.Hence if you care about the end result, go with Spacy. NLTK provides access to many algorithms. if you care about specific algo and customization go with NLTK

3. Spacy is More User friendly than NLTK

4.Spacy is best for APP Developers and NLTK is best for Researchers

```python
import spacy
```

Sentence Tokenization is Spacy

```python
nlp = spacy.load("en_core_web_sm")

doc = nlp("Dr.Strange loves to teach NLP. Hulk is his best friend.")

for sentence in doc.sents: #
    print(sentence)
```

```python
for sentence in doc.sents:
    for word in sentence:
        print(word)
```

```python
import nltk

nltk.download('punkt')
```

```python
from nltk.tokenize import sent_tokenize
sent_tokenize("Dr.Strange loves to teach NLP. Hulk is his best friend.")
```

```python
from nltk.tokenize import word_tokenize
word_tokenize("Dr.Strange loves to teach NLP. Hulk is his best friend.")
```
